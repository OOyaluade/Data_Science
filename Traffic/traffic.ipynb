```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python

traffic = pd.read_csv('traffic_sao_paulo.csv',sep=';')
```

```python
traffic.describe()
```

```python
traffic['Slowness in traffic (%)'] = traffic['Slowness in traffic (%)'].str.replace(',','.')
traffic['Slowness in traffic (%)'] = traffic['Slowness in traffic (%)'].astype(float)
```

```python
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
traffic_per_day = {}
plot_index  = 0
plt.figure(figsize=(10,12))
for i, day in zip(range(0, 135, 27), days):
    each_day_traffic = traffic[i:i+27]
    traffic_per_day[day] = each_day_traffic
    
    
    if plot_index < 6:

        plt.subplot(3,2,plot_index+1)
        plt.plot(each_day_traffic['Hour (Coded)'],each_day_traffic['Slowness in traffic (%)'])
        plt.title(day)
        plt.ylim(0,25)
    plot_index += 1

    
plt.subplot(3,2,6)
for day in days:
    day_of_week = traffic_per_day[day]
    plt.plot(day_of_week['Hour (Coded)'],day_of_week['Slowness in traffic (%)'],label=day)
    plt.ylim(0,25)
plt.legend()
plt.show()
    






    
```
