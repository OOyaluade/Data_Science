```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt, seaborn as sns
```

```python
housing = pd.read_csv('housing.csv')
```

```python
housing.head(5)
```

```python
housing.tail(5)
```

```python
housing.info()
```

```python
housing[['SalePrice','Gr Liv Area','Overall Qual','Garage Area','Rooms']]
```

```python
sns.set_theme()
sns.relplot(data=housing,x='Gr Liv Area',y='SalePrice',hue='Overall Qual',palette='RdYlGn',size='Garage Area',sizes=(1,300))
plt.show()
```

```python
sns.set_theme()

sns.relplot(data=housing,x='Gr Liv Area',y='SalePrice',hue='Overall Qual',palette='RdYlGn',size='Rooms',sizes=[200,50])
plt.show()
```

```python
sns.set_theme()
sns.relplot(data=housing,x='Gr Liv Area',y='SalePrice',hue='Overall Qual',palette='RdYlGn',size='Garage Area',sizes=(1,300),style='Rooms', markers=["*","D"])
plt.show()
```

```python
sns.relplot(data=housing, x='Gr Liv Area', y='SalePrice',
            hue='Overall Qual', palette='RdYlGn',
            size='Garage Area', sizes=(1, 300),
            style='Rooms', markers=['*','v'])
plt.show()
```

```python
sns.relplot(data=housing, x='Gr Liv Area', y='SalePrice',
            hue='Overall Qual', palette='RdYlGn',
            size='Garage Area', sizes=(1,300),
            style='Rooms', col='Rooms')
plt.show()
```
